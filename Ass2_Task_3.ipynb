```python
##Q2
import numpy as np #for nan

#show me the three colums
cols = ['BloodPressure', 'SkinThickness', 'BMI']
print(df[cols])

# number of '0' BEFORE
total_zeros = (df[cols] == 0).sum().sum()
print(f"Total '0': {total_zeros}")
#print(df['BloodPressure'].value_counts(ascending=False))

#replace '0' with null
df[cols] = df[cols].replace(0, np.nan)
print(df[cols])

# number of '0' AFTER
total_zeros = (df[cols] == 0).sum().sum()
print(f"Total '0': {total_zeros}")
```

```python
##Q3

for col in cols:
  for outcomes in [0, 1]:
    mean_col = df[df['Outcome'] == outcomes][col].mean()
    print("Mean for 'Outcome'", outcomes, "is", mean_col)

    check = df['Outcome'] == outcomes
    df.loc[check, col] = df.loc[check, col].fillna(mean_col)

print('We have NaN: ', df[cols].isna().sum())
```

```python
##Q4

sec_cor = df.corr()  #correlation WITH outcomes
sec_cor
```

```python
##Q5

from IPython.display import display #for better view, like a matrix

print('Correlation before:')
display(corr_mat)

print('COrrelation after:')
display(sec_cor)

result = sec_cor - corr_mat
print('The result is: ')
result
```
