```python
##Q1

#read our base
import pandas as pd
import numpy as np #for nan
from IPython.display import display

df = pd.read_csv('datasets/diabetes.csv')
```

```python

#making a new df without 'outcome' column
corr_mat = round(
    df.drop(
        columns='Outcome')
        .corr(),2
        )
corr_mat
```

```python
##Q2

#show me the three colums
cols = ['BloodPressure', 'SkinThickness', 'BMI']

for col in cols:
    print(f'Total 0 in column {col} before mod: {sum(df[col]==0)}\n')

df[cols] = df[cols].replace({0:np.nan})

for col in cols:
    print(f'Total 0 in column {col} after mod: {sum(df[col]==0)}\n')
```

```python
##Q3

for col in cols:
  for outcomes in [0, 1]:

    mean_col = round(df[df['Outcome'] == outcomes][col].mean(),2)
    print(f"\nMean for 'Outcome' {outcomes} for {col} is: ", mean_col)

    boolean = (df['Outcome'] == outcomes) & (df[col].isna())
    df.loc[boolean, col] = mean_col

print('\nWe have NaN: ', df[cols].isna().sum())
display(df)
```

```python
##Q4

sec_cor = round(df.corr(),5)  #correlation WITH outcomes
sec_cor
```

```python
##Q5

 #for better view, like a matrix

print('Correlation before:')
display(corr_mat)

print('COrrelation after:')
display(sec_cor)

result = sec_cor - corr_mat
print('The result is: ')
result
```
